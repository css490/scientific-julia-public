```julia
#The sigmoid function
function sigmoid(z)
    g = 1 ./ (1 + e.^(-z))
end
```

```julia
#The first two columns contain scores on two exams respectively,
#the third column contains the labels for admitted or not admitted
data = readcsv("admissions.txt")
```

```julia
#X is the matrix of grades on exams 1 and 2
X = data[:,1:2]

#y is the true outcome
y = data[:,3]
```

```julia
function plotData(X, y)
  pos = find(y)
  neg = find(abs(y-1))
  plot(X[pos,1], X[pos,2], "+")
  plot(X[neg,1], X[neg,2], "o")

  xlabel("Exam 1 score")
  ylabel("Exam 2 score")
  legend(("Admitted", "Not admitted"), "upper right", numpoints=1)
end    

plotData(X, y)
```

```julia
m, n = size(X)
X = [ones(m,1) X] #add the ones feature to x
```

```julia
function cost(X, y, theta)
    m = length(y)  
    C = 0 #cost
    for i=1:m
        C += (1/m)*(-y[i]*log(sigmoid(X[i,:]*theta)) - (1 - y[i])*log(1-sigmoid(X[i,:]*theta)))
    end
    return C
end

function costGrad(X, y, theta)
    m = length(y)  
    delta = zeros(size(theta)) #partial derivatives with respect to each theta
    for i=1:m
        delta += (1/m)*(sigmoid(X[i,:]*theta) - y[i])[1]*(X[i,:])'
    end
    return delta
end
```

```julia
init_theta = zeros(size(X,2),1)
C, delta = cost(X, y, init_theta), costGrad(X, y, init_theta)
```

```julia
f(theta::Vector) = cost(X, y, reshape(theta,3,1))[1]
function g!(theta::Vector, storage::Vector) 
   storage[1] = costGrad(X, y, reshape(theta,3,1))[1]
   storage[2] = costGrad(X, y, reshape(theta,3,1))[2]
   storage[3] = costGrad(X, y, reshape(theta,3,1))[3]
end

res = optimize(f, g!, [0.0, 0.0, 0.0])
```

```julia
theta = res.minimum
#Plotting the decision boundary
#Only need 2 points to define a line, so choose two endpoints
plot_x = [minimum(X[:,2])-2,  maximum(X[:,2])+2];

#Calculate the decision boundary line
plot_y = (-1./theta[3]).*(theta[2].*plot_x + theta[1]);

#Plot the data
plotData(X[:,2:3],y)
#Plot the decision boundary, and adjust axes for better viewing
plot(plot_x, plot_y)
```

```julia
function predict(theta, X)
    m = size(X,1)
    p = zeros(m,1)
    
    for i = 1:m
        if (sigmoid(X[i,:]*theta)[1] >= 0.5)
            p[i] = 1
      else 
            p[i] = 0 
      end
    end
    return p
end    
```

```julia
function gradientDescent(X, y, theta, alpha, iter)
    m = length(y)
    Jhist = zeros(iter,1)
    for k=1:iter
        delta = zeros(size(theta)) #partial derivatives with respect to each theta
        for i=1:m
            delta += (1/m)*(sigmoid(X[i,:]*theta) - y[i])[1]*(X[i,:])'
        end
        theta = theta - (alpha/m)*delta
        Jhist[k] = cost(X, y, theta)[1]
    end
    return theta, Jhist
end
```

```julia
iterations = 1000
alpha = 0.1
init_theta = [0.0; 0.0; 0.0]

result = gradientDescent(X, y, init_theta, alpha, iterations)
```

```julia
plot(log(1:length(result[2])), log(result[2]), color="red")
xlabel("Iterations"); ylabel("Cost")
```
