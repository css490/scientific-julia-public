```julia
#sample call jacobi_heat(120, 100000)
function jacobi_heat(N, maxiter)
    #grid spacing
    h = 1 / (N + 1)
    
    #tolerance
    tolerance = 0.1 / (N+1)^2
    
    #x ad y coordinates of grid points
    x = [0:h:N+1]
    y = [0:h:N+1]
    
    #initialize U
    U = zeros(N+2, N+2)
    
    #top and right boundaries should be initialized to 1
    #U[:, N+2] = ones(N+2)
    #U[N+2, :] = ones(N+2)'
    
    for k = 1:N+2
        if 0.3 < x[k] < 0.7
            U[k, N+2] = 1
        end
        
        if 0.3 < y[k] < 0.7  
            U[N+2, k] = 1
        end
    end
            
    
    iter = 0
    dumax = tolerance
    
    while iter < maxiter && dumax >= tolerance
        Uold = U 
        for j = 2:N+1
            for i = 2:N+1
                U[i,j] = (1/4)*(Uold[i-1,j] + Uold[i+1,j] +
                          Uold[i,j-1] + Uold[i,j+1])
                dumax = max(dumax, abs(U[i,j] - Uold[i,j]))
            end
        end
        iter += 1
    end
    return U
end
```

```julia
using PyPlot
```

```julia
U = jacobi_heat(120, 100000)
```

```julia
figure(1)                  
clf()                       
axis("equal")
pcolormesh(U)
title("Temperature")
#savefig("pcolor.png")

figure(2)
clf()
axis("equal")
contour(U)
#savefig("contour.png")
```
